# Clean and subset dataset

This script removes noisy, unsure, uncertain, hybrid and nonfocal calls from the dataset, as well as extremely long and short calls.
It also creates a completely balanced subset (equal N of each calltype) and a reduced subset from the original dataset. The reduced subset will be used for all further analyses. It has reduced N, but mostly the same calltype frequencies as in orginal dataset. 
Only cc call frequency is reduced as they are by far the most frequent calltype and overshadow every other call.

### Clean dataset 

#### Removing unsure, uncertain, hybrid calls etc

```python
print("Full: ", df.shape)
df = df.loc[df['unsure_call']==0]
print("Removed unsure: ", df.shape)

df = df.loc[df['certain_yn']=="y"]
print("Removed uncertain: ",df.shape)

df = df.loc[df['noisy']==0]
print("Removed noisy: ",df.shape)

df = df.loc[df.call_lable != 'fu/hyb',:]
print("Removed hybrids: ",df.shape)

print("Removed nonfocal: ",df.shape)
df = df.loc[df['nonfocal_yn']=="n"]
```

#### Create a cc reduced dataset

```python
wo_cc_df = df[df.call_lable != 'cc']
cc_df = df[df.call_lable == 'cc']
cc_subset = cc_df.sample(frac=0.1, replace=False, random_state=1)
df = pd.concat((wo_cc_df, cc_subset), ignore_index=False, axis=0)

dftype='reduced'
```

#### Create a completely balanced dataset

```python
labelcounts = pd.DataFrame(df['call_lable'].value_counts())
class_size = np.min(labelcounts.call_lable.values)

subsets = []
for label, num in zip(labelcounts.index, labelcounts.call_lable.values):
    print(label, ":", num, ":", class_size/num)
    subsets.append(df[df.call_lable == label].sample(frac=(class_size/num), replace=False, random_state=1))

df = pd.concat(subsets, axis=0)
dftype = 'balanced'
```

```python
labelcounts = pd.DataFrame(df['call_lable'].value_counts())
labelcounts.sort_index(inplace=True)
pal = sns.color_palette("Set2", n_colors=len(labelcounts.index))
plt.barh(labelcounts.index, labelcounts.call_lable.values, color = pal)

for index, value in enumerate(labelcounts.call_lable.values):
    plt.text(value, index, str(value))

plt.title(dftype.capitalize()+' dataset (N='+str(df.shape[0])+')')
plt.ylabel('Labels')
plt.xlabel('Frequencies')
plt.xlim(0,21500)
#plt.show()
outpath = os.path.join(os.path.sep, FIGURES, "labelcounts_"+dftype+".png")
plt.savefig(outpath, facecolor='w')
print("Saved as: ", outpath)
```

#### Removing duplicates

There are 6 duplicates in the full dataset, 2 in the reduced  and none in the balanced, but I did not realize until the very end of the analysis. Therefore, I'm adding the remove-duplicates step here, so that the composition of subsets will not change.

##### Reduced dataset

```python
dftype = "reduced"

df = pd.read_pickle(os.path.join(os.path.sep, DATA, "df_focal_"+dftype+".pkl"))
print(df.shape)
```

```python
all_ids = list(df.callID.values)
duplics = [item for item, count in collections.Counter(all_ids).items() if count>1]
print(len(duplics), " duplicates found")
df[df['callID'].isin(duplics)].sort_values(by=['callID'])
```

```python
df = df.drop_duplicates(subset='callID', keep="last")
print(df.shape)
df.to_pickle(os.path.join(os.path.sep, DATA, "df_focal_"+dftype+".pkl"))
```

```python
labelcounts = pd.DataFrame(df['call_lable'].value_counts())
labelcounts.sort_index(inplace=True)
pal = sns.color_palette("Set2", n_colors=len(labelcounts.index))
plt.barh(labelcounts.index, labelcounts.call_lable.values, color = pal)

for index, value in enumerate(labelcounts.call_lable.values):
    plt.text(value, index, str(value))

plt.title(dftype.capitalize()+' dataset (N='+str(df.shape[0])+')')
plt.ylabel('Labels')
plt.xlabel('Frequencies')
plt.xlim(0,21500)

outpath = os.path.join(os.path.sep, FIGURES, "labelcounts_"+dftype+".png")
plt.savefig(outpath, facecolor='w')
print("Saved as: ", outpath)
```
