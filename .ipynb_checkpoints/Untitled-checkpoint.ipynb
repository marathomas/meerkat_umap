```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
from matplotlib.legend import Legend
import matplotlib
import seaborn as sns
import plotly.express as px
import plotly.graph_objects as go
import random


def umap_2Dplot(x,y, scat_labels, mycolors, outname=None, showlegend=True, **kwargs):
    
    labeltypes = sorted(list(set(scat_labels)))
    pal = sns.color_palette(mycolors, n_colors=len(labeltypes))
    color_dict = dict(zip(labeltypes, pal))
    c = [color_dict[val] for val in scat_labels]
    
    fig = plt.figure(figsize=(6,6))
    
    plt.scatter(x, y, alpha=1,
                c=c, **kwargs)
    plt.xlabel('UMAP1')
    plt.ylabel('UMAP2');

    scatters = []
    for label in labeltypes:
        scatters.append(matplotlib.lines.Line2D([0],[0], linestyle="none", c=color_dict[label], marker = 'o'))
    
    if showlegend: plt.legend(scatters, labeltypes, numpoints = 1) 
    if outname: plt.savefig(outname, facecolor="white")
```

```python
x = np.random.rand(40,)
y = np.random.rand(40,)
labeltypes = ['orange', 'banana', 'apple', 'pear']
labels = random.choices(labeltypes, weights=[1,3,1,2],k=40)
```

```python
umap_2Dplot(x,y,labels, "Set2")
```

```python
umap_2Dplot(x,y,labels, "Set2", s=50, marker="x")
```
